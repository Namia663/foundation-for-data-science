Nomor 1.5 Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
Normal: 18.5-24.9
Overweight: 25.0-29.9
Obesity: 30.0-34.9
Underweight: < 18.5


```python
status_list=[]
for i in range (len(df)):
  if 18.5 <= df['BMI'][i] <= 24.9:
    status = 'normal'
  elif 25.0 <= df['BMI'][i] <= 29.9:
    status = 'overweight'
  elif 30 <= df["BMI"][i] <= 34.9:
    status = 'obesity'
  elif df["BMI"][i] < 18.5:
    status = 'underweight'
  status_list.append(status)
df['Status BMI']=status_list
df
```

Nomor 1.9

Cari nilai-nilai berikut:
- Cari nilai terendah dan tertingginya.
- Berapa nilai rata-rata tinggi badan siswa?
- Berapa rata-rata berat badan siswa?
- Ada berapa orang jumlah siswa yang tercatat pada dataframe df?
- 5 orang dengan nilai BMI tertinggi.
- Ada berapa orang yang tergolong status Underweight, Normal, Overweight,
dan Obesity?

```python
import pandas as pd

# Cari nilai terendah dan tertingginya (diasumsikan nilai BMI)
nilai_terendah = df['BMI'].min()
nilai_tertinggi = df['BMI'].max()

# Cari nilai rata-rata tinggi badan dan berat badan siswa
rata_rata_tinggi_badan = df['Tinggi Badan (cm)'].mean()
rata_rata_berat_badan = df['Berat Badan (kg)'].mean()

# Hitung jumlah siswa
jumlah_siswa = len(df)

# Temukan 5 orang dengan nilai BMI tertinggi
df_sort_bmi_tertinggi = df.sort_values(by='BMI', ascending=False).head(5)

# Hitung jumlah siswa berdasarkan status BMI
jumlah_underweight = df[df['Status BMI'] == 'underweight'].shape[0]
jumlah_normal = df[df['Status BMI'] == 'normal'].shape[0]
jumlah_overweight = df[df['Status BMI'] == 'overweight'].shape[0]
jumlah_obesity = df[df['Status BMI'] == 'obesity'].shape[0]

# Menampilkan hasil
print("Nilai terendah BMI:")
print(nilai_terendah)
print("\nNilai tertinggi BMI:")
print(nilai_tertinggi)
print("\nRata-rata tinggi badan siswa:", rata_rata_tinggi_badan, 'cm')
print("Rata-rata berat badan siswa:", rata_rata_berat_badan, 'kg')
print("\nJumlah siswa:", jumlah_siswa)
print("\n5 orang dengan nilai BMI tertinggi:")
print(df_sort_bmi_tertinggi)
print("\nJumlah siswa berdasarkan status BMI:")
print("Underweight:", jumlah_underweight)
print("Normal Weight:", jumlah_normal)
print("Overweight:", jumlah_overweight)
print("Obesity:", jumlah_obesity)
```

##**NOMOR 2**

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.


```python
def calculate_investment(capital, tenure):
    # Validasi input
    if capital < 0 or capital > 1000000000:
        print("Error: Modal investasi harus dalam rentang antara 0 hingga 1.000.000.000.")
        return
    elif tenure < 1 or tenure > 25:
        print("Error: Jangka waktu investasi harus dalam rentang antara 1 hingga 25 tahun.")
        return

    # Inisialisasi variabel
    total_investment = capital

    # Menghitung akumulasi investasi setiap tahun
    for year in range(1, tenure + 1):
        total_investment *= 1.25  # Menambahkan keuntungan 25% setiap tahun
        print('Year ', year, ', your total investment: ', total_investment)


# Input modal investasi dan jangka waktu investasi dari pengguna
try:
    capital = int(input("Money Capital: "))
    tenure = int(input("Investment Tenure (years): "))
    calculate_investment(capital, tenure)
except ValueError:
    print("Error: Input harus berupa bilangan bulat.")
```
